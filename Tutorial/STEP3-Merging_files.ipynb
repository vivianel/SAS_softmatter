# Merge files

```python
import matplotlib.pyplot as plt    # Load matplotlib
import numpy as np
import os
```

# Check the file merging with GC as the absolute standard
## Load all files from 42 mm

```python
# all edf data is colleced in a folder with the same detector distance
folder_42 = 'integrated_GC_42mm'

# now we load all the files in that folder
files_42 = []
# r=root, d=directories, f = files
for r, d, f in os.walk(folder_42):
    for file in f:
        if '.dat' in file:
            files_42.append(os.path.join(file))
# %% LIST NAMES
for ii in range(len(files_42)):
   print(ii, '|', files_42[ii])
```

## Load all files from 900 mm

```python
# all edf data is colleced in a folder with the same detector distance
folder_900 = 'integrated_GC_900mm'

# now we load all the files in that folder
files_900 = []
# r=root, d=directories, f = files
for r, d, f in os.walk(folder_900):
    for file in f:
        if '.dat' in file:
            files_900.append(os.path.join(file))
# %% LIST NAMES
for ii in range(len(files_900)):
   print(ii, '|', files_900[ii])
```

## Plot all the files together to judge the fitting

```python
merged_folder = 'merged_GC'
cut_900mm = -150
cut_42mm = 30

for ii in range(len(files_42)):
    f_42 = np.loadtxt(folder_42 +'/'+ files_42[ii])
    f_900 = np.loadtxt(folder_900 +'/'+ files_900[ii])
       # concatenate the file
    q = np.concatenate((f_900[:cut_900mm,0], f_42[cut_42mm:,0]))
    I = np.concatenate((f_900[:cut_900mm:,1], f_42[cut_42mm:,1]))
    # order the array
    idx = np.argsort(q)
    q = q[idx]
    I = I[idx]
    # plot the merged data
    #plt.figure()
    plt.loglog(f_900[:cut_900mm,0], f_900[:cut_900mm,1])
    plt.loglog(f_42[cut_42mm:,0], f_42[cut_42mm:,1])
    plt.title([files_900[ii], files_42[ii]])
    # settings for the plot
    plt.xlabel('Scattering vector q [1/A]')
    plt.ylabel('Differential Scattering cross section [1/cm]')
    plt.grid(which='both', linestyle='--', linewidth=0.5)
    plt.loglog(q, I)
    # save the data
    save_as = merged_folder + '/' + files_42[ii][:-4] + '.dat'
    save_data = np.transpose(np.vstack([q, I]))
    np.savetxt(save_as , save_data)
```
